```python
import numpy as np
import pandas as pd

from sklearn import linear_model as lm
from sklearn import preprocessing as pp
from sklearn import pipeline as skp



fp = open('/Users/dad/Downloads/input.txt')
F, N = map(int, fp.readline().split())
train = np.array([fp.readline().split() for _ in range(N)], float)
T = int(fp.readline())
test = np.array([fp.readline().split() for _ in range(T)], float)
```

```python
def PolynomialRegression(degree=2,**kwargs):
    return skp.make_pipeline(pp.PolynomialFeatures(degree), lm.LinearRegression(**kwargs))
model = PolynomialRegression()
model.fit(train[:,:-1], train[:,-1])
```

```python
from sklearn import metrics
ypred = model.predict(train[:,:-1])
metrics.mean_absolute_error(train[:,-1], ypred)
```
